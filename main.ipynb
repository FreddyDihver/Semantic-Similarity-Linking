```python
from SentenceFunctions import *
```

## Setup

```python
# Choose pre-trained model
model = SentenceTransformer("all-MiniLM-L6-v2")
# New name after training
new_model_name = "Models/_"
```

```python
# Load data
Nsamples = 5000
path = "path/to/training/data/csv/file"
ori, tar, labels, ori_eval, tar_eval, labels_eval = load_bal_train_unbal_eval(path, Nsamples=Nsamples, trainsplit=0.7)
```

#### Build structure

```python
# Choose builder function
bf = build_token_structure_with_support

# Build structure
ori_sentence, tar_sentence = build_structure(ori, tar, bf, model=model)
```

# Training

```python
# Prepare training data
train_dataset = prepare_training(ori_sentence, tar_sentence, labels)
lossfn = losses.CosineSimilarityLoss
```

```python
# Train model (default loss: CosineSimilarityLoss)
num_epochs = 2
train_model(train_dataset, model = model, num_epochs = num_epochs, loss_fn=lossfn)
```

# Post-process

```python
# Self validation
F1, prec, rec, predmat, thres = F1_fn(model, ori_sentence, tar_sentence, labels, maximize_F1_prec=True)[:5]
```

#### Evaluation

```python
# Build eval sentences
ori_eval_sentence, tar_eval_sentence = build_structure(ori_eval, tar_eval, bf, model=model)
labels_eval = ori_eval["label"].values
```

```python
# Test metrics
F1_fn(model, ori_eval_sentence, tar_eval_sentence, labels_eval, thres, maximize_F1_prec=True);
```

```python
model.save(new_model_name)
```
